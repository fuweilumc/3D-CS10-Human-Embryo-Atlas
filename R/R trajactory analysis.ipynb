# 2. for nural tube (extended data fig.4 i,j)

```python
seu_obj= schard::h5ad2seurat('nt_clean_counts_only.h5ad')
```

```python
cds <- as.cell_data_set(seu_obj)
cds <- preprocess_cds(cds, num_dim = 30)

cds <- reduce_dimension(cds, 
                        preprocess_method = "PCA",
                        umap.min_dist = 0.1,      
                        umap.n_neighbors = 15)

plot_cells(cds, color_cells_by = "leiden_hvg_1", label_groups_by_cluster = TRUE,
           label_leaves = FALSE, label_branch_points = FALSE, cell_size = 1,label_cell_groups =TRUE)+theme(legend.position = "right")
```

```python
cds <- cluster_cells(cds,k=15)
plot_cells(cds, color_cells_by = "partition") 
```

```python
sub_cluster_palette <- c(
    "e0_0" = "#ff002a",
    "e0_1" = "#000080",
    "e0_2" = "#e5e022",
    "e0_3" = "#009203",
    "e0_4" = "#a8329e",
    "e0_5" = "#00FFFF",

    "e12_0" = "#FF8C00",
    "e12_1" = "#565DFD",
    "e12_2" = "#FF00FF",
    "e12_3" = "#8B4513",

    "e20_0" = "#66c102",
    "e20_1" = "#8e0119",
    "e20_2" = "#10686f",

    "e25_0" = "#FF1493",
    "e25_1" = "#9f50f9",
    "e25_2" = "#32a88e",

    "0" = "#FF1493",
    "1" = "#9f50f9",
    "2" = "#32a88e"
)
```
